```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../../main/sub_data/single_variant/sin_ndlc_psp.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
null_dates = []
for i in range(len(df)):
    if pd.isnull(df.iloc[i][2]):
        null_dates.append(df.iloc[i][0])
```

```python
null_dates
```

## Interpolation Imputation

    A type of Imputation which is common way to deal with missing values

```python
missing = df.isna()
df_interp_linear = df.interpolate('linear')
ax = df_interp_linear.plot(c='r')
df.plot(ax=ax)
```

```python
df.iloc[60]
```

```python
def check(data):
    res = []
    for i in null_dates:
        res.append(data.iloc[i][2])
    return pd.DataFrame(res)
```

```python
missing = df.isna()
df_interp_time = df.interpolate('index')
ax = df_interp_time.plot(c='r')
df.plot(ax=ax)
(check(df_interp_time) == check(df_interp_linear)).sum() == 73
```

```python
missing = df.isna()
df_interp_val = df.interpolate('values')
ax = df_interp_val.plot(c='r')
df.plot(ax=ax)
(check(df_interp_val) == check(df_interp_linear)).sum() == 73
```

```python
missing = df.isna()
df_interp_near = df.interpolate('nearest')
ax = df_interp_near.plot(c='r')
df.plot(ax=ax)
(check(df_interp_near) == check(df_interp_linear)).value_counts()
```

```python
missing = df.isna()
df_interp_zero = df.interpolate('zero')
ax = df_interp_zero.plot(c='r')
df.plot(ax=ax)
print((check(df_interp_zero) == check(df_interp_near)).value_counts())
print((check(df_interp_zero) == check(df_interp_linear)).value_counts())
```

```python
missing = df.isna()
df_interp_slinear = df.interpolate('slinear')
ax = df_interp_slinear.plot(c='r')
df.plot(ax=ax)
print((check(df_interp_slinear) == check(df_interp_near)).value_counts())
print((check(df_interp_slinear) == check(df_interp_linear)).value_counts())
```

```python
missing = df.isna()
df_interp_deri = df.interpolate('from_derivatives')
ax = df_interp_deri.plot(c='r')
df.plot(ax=ax)
print((check(df_interp_deri) == check(df_interp_near)).value_counts())
print((check(df_interp_deri) == check(df_interp_linear)).value_counts())
print((check(df_interp_deri) == check(df_interp_slinear)).value_counts())
```

```python
df1 = df_interp_linear.drop("Unnamed: 0",axis=1)
df2 = df_interp_slinear.drop("Unnamed: 0",axis=1)
df3 = df_interp_zero.drop("Unnamed: 0",axis=1)
df4 = df_interp_deri.drop("Unnamed: 0",axis=1)
df5 = df_interp_near.drop("Unnamed: 0",axis=1)
```

```python
df1.to_csv("../../../main/sub_data/single_variant/interpolation/sin_linear.csv")
df2.to_csv("../../../main/sub_data/single_variant/interpolation/sin_slinear.csv")
df3.to_csv("../../../main/sub_data/single_variant/interpolation/sin_zero.csv")
df4.to_csv("../../../main/sub_data/single_variant/interpolation/sin_from_derivatives.csv")
df5.to_csv("../../../main/sub_data/single_variant/interpolation/sin_nearest.csv")
```
