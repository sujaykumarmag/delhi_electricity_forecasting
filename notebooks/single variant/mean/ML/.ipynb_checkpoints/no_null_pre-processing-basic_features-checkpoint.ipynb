```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error as MAPE
```

```python
df = pd.read_csv("../../../main/sub_data/single_variant/no_null/train/train_no_null.csv",index_col="Date",parse_dates=True)
df_test = pd.read_csv("../../../main/sub_data/single_variant/no_null/test/test_no_null.csv",index_col="Date",parse_dates=True)
```

```python
df
```

## Transforming Data
    
    The Steps adopted for Transformation are
<ul>
    <li>Rolling Window</li>
    <li>Outliers Replacement</li>
</ul>

```python
# We use Aggregate method (for this, we need a function)
def percent_change(series):
    prev = series[:-1]
    last = series[-1:]
    change = (last-np.sum(prev))/np.sum(prev)
    return change

def replace_outliers(series):
    mean_diff = np.abs(series - np.mean(series))
    # Criteria for detecting outliers
    this_mask = mean_diff > (np.std(series) * 3)
    # replace it with median
    series[this_mask] = np.nanmedian(series)
    return series
```

```python
'''
    Rolling : A rolling window of every size (n) is applied to every column
    After applying an n size NaN Values will be available
    
'''
df_per_ch = df.rolling(20).aggregate(percent_change).dropna()
```

```python
df_per_ch
```

```python
# Now Visualize it
df_per_ch.loc[:"2014"].plot()
plt.show()
```

```python
df_per_ch = df_per_ch.apply(lambda col : replace_outliers(col))
```

```python
# Now Visualize it
df_per_ch.loc[:"2014"].plot()
plt.show()
```

```python
features_basic = df_per_ch.rolling(20,min_periods=5, closed='right').aggregate([np.max,np.min,np.mean,np.std])
```

```python
features_basic = features_basic.dropna()
```

```python
features_basic["demand"] = df_per_ch["Max.Demand met during the day (MW)"]
```

```python
cols = features_basic.columns
X = features_basic[cols[0:4]]
y = features_basic[cols[4:]]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
# Fit the Ridge model
model = Ridge()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
MAPE(y_test,y_pred)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
y_pred = lasso.predict(X_test)
MAPE(y_test,y_pred)
```

```python
from prophet import Prophet
model = Prophet()
df = pd.read_csv("../../../main/sub_data/single_variant/no_null/train/train_no_null.csv")
df.columns = ['ds','y']
model.fit(df)
future_dates = model.make_future_dataframe(periods=31)[3609:]
forecast = model.predict(future_dates)
MAPE(forecast['trend'],df_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbm = GradientBoostingRegressor()
gbm.fit(X_train, y_train)
y_pred = gbm.predict(X_test)
MAPE(y_pred,y_test)
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svr = SVR()
svr.fit(X_train_scaled, y_train)
y_pred = svr.predict(X_test_scaled)
MAPE(y_test,y_pred)
```
