```python
import os
```

```python
folder_path = "test/"
output_path = "processed/"
files = os.listdir(folder_path)
```

# First Convert to TXT

```python
def pdf_conversion(file):
    org_path = folder_path+file
    processed_path = output_path+file[0:8]+".txt"
    pdffileobj=open(org_path,'rb')
    pdfreader=PyPDF2.PdfReader(pdffileobj)
    x=len(pdfreader.pages)
    pageobj=pdfreader.pages[1]
    text=pageobj.extract_text()
    # Make a Txt File
    file1=open(processed_path,"a")
    file1.writelines(text)
    file1.close()
    pdffileobj.close()
```

```python
for file in files:
    pdf_conversion(file)
```

# Extract Info from the files

```python
def extract_info(file):
    file_path = output_path+file
    myfile = open(file_path, "rt") 
    contents = myfile.read()       
    myfile.close()                 
    if "Delhi" in contents:
        x = contents.index("Delhi")
    print(x)
    stri = ""
    s = []
    s.append(file[0:8])
    sd = ""
    while stri != "\n" and x<len(contents):
        if stri == " ":
            s.append(sd)
            sd = contents[x]
        else:
            sd = sd + contents[x]
        stri = contents[x]
        x = x+1
    sd = sd[:4]
    s.append(sd)
    return s
```

```python
processed_files = os.listdir(output_path)
```

```python
data = []
for file in processed_files:
    data.append(extract_info(file))
```
