```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("../../../main/ndlc_psp_final.csv")
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df[i].isnull().sum())
```

```python
df = df.dropna(subset=['Max OD (MW)'])
```

```python
def regression_imputation(dataframe, variable):
    
    # Create a copy of the DataFrame
    df_copy = dataframe.copy()
    df_copy = df_copy.drop(["Date (DD/MM/YYYY)","Unnamed: 0"],axis=1)
    train_data = df_copy[df_copy[variable].notnull()]
    
    # Create a DataFrame with missing values for the target variable
    test_data = df_copy[df_copy[variable].isnull()]
    
    # Separate features and target variables for training and testing data
    X_train = train_data.drop(variable, axis=1)
    y_train = train_data[variable]
    X_test = test_data.drop(variable, axis=1)
    
    # Build a linear regression model
    regression_model = LinearRegression()
    
    # Fit the model on the training data
    regression_model.fit(X_train, y_train)
    
    # Predict missing values based on the model
    imputed_values = regression_model.predict(X_test)
    
    # Assign the predicted values to the missing values in the DataFrame
    dataframe.loc[dataframe[variable].isnull(), variable] = imputed_values
```

```python
regression_imputation(df, "Energy Shortage (MU)")
```
