```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("../main/ndlc_psp_final.csv")
```

```python
df
```

```python
df['Date (DD/MM/YYYY)'] = pd.to_datetime(df['Date (DD/MM/YYYY)'])
```

```python
df.dtypes
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df[i].isnull().sum())
```

```python
df = df.dropna(subset=['Max OD (MW)'])
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df[i].isnull().sum())
```

```python
df_mean = df.fillna(df["Energy Shortage (MU)"].mean())
df_median = df.fillna(df["Energy Shortage (MU)"].median())
df_mode = df.fillna(df["Energy Shortage (MU)"].mode().iloc[0])
```

```python
(df_mode == df_median).sum() == len(df)
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df_mean[i].isnull().sum())

print("\n=======================================================================================\n")
    
for i in cols:
    print(i,'-------------------',df_median[i].isnull().sum())

print("\n=======================================================================================\n")
    
for i in cols:
    print(i,'-------------------',df_mode[i].isnull().sum())
```
