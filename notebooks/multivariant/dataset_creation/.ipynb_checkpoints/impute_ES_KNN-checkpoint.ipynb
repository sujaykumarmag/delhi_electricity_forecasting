```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv("../../../main/ndlc_psp_final1.csv")
```

```python
df
```

```python
df.columns = ['Date (DD/MM/YYYY)', 'Unnamed: 0', 'Max.Demand met during the day (MW)',
       'Shortage during maximum Demand (MW)', 'Energy Met (MU)',
       'Drawal Schedule\n(MU)', 'OD(+)/UD(-) (MU)', 'Max OD (MW)',
       'Energy Shortage (MU)']
```

```python
df = df.dropna(subset=['Max OD (MW)'])
df['Date (DD/MM/YYYY)'] = pd.to_datetime(df['Date (DD/MM/YYYY)'])
```

```python
# Convert dates to numerical representation (elapsed time in days)
df['Date (DD/MM/YYYY)'] = (df['Date (DD/MM/YYYY)'] - pd.to_datetime('2000-01-01')).dt.days
```

```python
df
```

```python
df.dtypes
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df[i].isnull().sum())
```

```python
imputer = KNNImputer(n_neighbors=3) 
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df_imputed
```

```python
df_imputed['Date (DD/MM/YYYY)'] = pd.to_datetime('2000-01-01') + pd.to_timedelta(df_imputed['Date (DD/MM/YYYY)'], unit='D')
```

```python
df_imputed
```

```python
cols = df.columns
for i in cols:
    print(i,'-------------------',df_imputed[i].isnull().sum())
```

```python
df_imputed
```

```python
df_imputed.to_csv("../../../main/sub_data/multi_variant/ndlc_psp_ES_KNN.csv")
```
